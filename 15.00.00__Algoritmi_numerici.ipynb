# Metodo di bisezione per trovare lo zero di una funzione

Assumiamo che una funzione sia monotona in un intervallo.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Si parte dai due estremi; si verifica che il segno della funzione sia opposto, e si divide in due l'intervallo. Poi si scarta l'intervallo $[x_1, x_m]$ oppure $[x_m, x_2]$ dove la funzione non interseca $y=0$.

```python
def root(f, x1, x2, xtol=1e-5, maxiter=100):
    f1, f2 = f(x1), f(x2)
    if f(x1)*f(x2)>0:
        print('error: y=f(x) does not intersect y=0')
        return 0
    iters = 0
    pts = []
    while (x2 - x1) > xtol and iters < maxiter:
        xm = (x1+x2)/2
        fm = f(xm)
        if f1*fm> 0: x1, f1 = xm, fm
        else: x2, f2 = xm, fm
        iters += 1
        print(iters, x1, x2)
        pts.append(xm)
    return (x1+x2)/2, np.array(pts)
```

```python
def f(x):
    return x**3*np.exp(-3*x)-0.02
r, pts = root(f, 0, 1)
print("x0 =", r)
plt.figure(figsize=(8,6), dpi=100)
xs = np.linspace(0, 1, 100)
ys = f(xs)
plt.plot(xs, ys)
plt.plot(xs, np.zeros(len(xs)), color='darkred', linestyle='--')
plt.scatter(pts, f(pts), color='darkblue', s=10)
plt.scatter([r], [0], color='darkred', s=50)
plt.show()
```

# Ricerca del minimo di una funzione: *gradient descent*

Assumiamo che una funzione sia monotona in un intervallo.

```python
def fmin(f, f1, x0, rate=0.1, xtol=0.000001, maxiter=1000):
    delta = 1
    iters = 0
    x = x0
    pts = [x0]
    while delta > xtol and iters < maxiter:
        x0 = x
        x -= rate*f1(x0)
        delta = abs(x - x0)
        iters += 1
        print(iters, x)
        pts.append(x)
    return x, np.array(pts)
```

```python
def f(x):
    return 10*(x-1.78)**2 - 0.7*x**3
def f1(x):
    return 10*2*(x-1.78) - 0.7*3*x**2
plt.figure(figsize=(8,6), dpi=100)
xs = np.linspace(0, 6, 100)
ys = f(xs)
xmin, pts = fmin(f, f1, 5, 0.07)
plt.plot(xs, ys)
plt.scatter(pts, f(pts), color='darkblue', s=10)
plt.scatter([xmin], [f(xmin)], color='darkred', s=50)
plt.show()
```

Se la derivata della funzione non  nota, dobbiamo cavarcela da soli. Dobbiamo per fare attenzione alla stabilit numerica dell'algoritmo che usiamo per calcolare la derivata!

```python
def fmin1(f, f1, x0, rate= 0.1, xtol=0.000001, maxiter=1000):
    delta = 1
    iters = 0
    x = x0
    pts = [x0]
    while delta > xtol and iters < maxiter:
        x0 = x
        f1 = (f(x0+xtol)-f(x0))/xtol
        x -= rate*f1
        delta = abs(x - x0)
        iters += 1
        pts.append(x)
        print(iters, x)
    return x, np.array(pts)
```

```python
def f(x):
    return 10*(x-1.78)**2 - 0.7*x**3
plt.figure(figsize=(8,6), dpi=100)
xs = np.linspace(0, 6, 100)
ys = f(xs)
xmin, pts = fmin1(f, f1, 5, 0.07)
plt.plot(xs, ys)
plt.scatter(pts, f(pts), color='darkblue', s=10)
plt.scatter([xmin], [f(xmin)], color='darkred', s=50)
plt.show()
```

# Minimizzazioni a pi variabili

Lo stesso metodo pu essere applicato seguento il **gradiente** della funzione, che generalizza la derivata nel caso monodimensionale.

In pi dimensioni, per funzioni complesse, pu essere problematica la ricerca del **minimo assoluto** quando sono presenti dei **minimi locali** nei quali l'algoritmo pu rimanere bloccato.

Per questo, pu essere utile privare a cambiare il valore inziale dal quale parte la ricerca del minimo e verificare qual  il minimo raggiunto.

Nei problemi di fit, spesso la natura della funzione che modellizza il problema ci potr aiutare. Ad esempio, se cerchiamo un picco, un valore iniziale vicino al centro del picco ci aiuter a far convergere il fit verso il valore ottimale.
